```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
years = range(2021, 2026)
total_posts = [19,  148, 1456, 1208,42]
total_users = [12, 123, 1422, 996,72]

df = pd.DataFrame({
    'Year': years,
    'TotalPosts': total_posts,
    'TotalUsers': total_users
})


bar_width = 0.1  
group_spacing = 0.3  
x = np.arange(len(years)) * group_spacing

colors = ['#0f4fdb', '#6f98ec']  

plt.rcParams.update({
    'font.size': 12,
    'font.family': 'Times New Roman',  # 数字和英文使用Times New Roman
    'axes.unicode_minus': False        # 解决负号显示问题
})
# 创建画布
plt.figure(figsize=(6, 4))

# 绘制双柱状图
bars1 = plt.bar(x - bar_width*0.4, df['TotalPosts'], width=bar_width/1.5,
                color=colors[0], label='TotalPosts')
bars2 = plt.bar(x + bar_width*0.4, df['TotalUsers'], width=bar_width/1.5,
                color=colors[1], label='TotalUsers')

# 添加标签和样式
plt.xticks(x, df['Year'])
plt.yticks(np.arange(0, 1501, 500))
plt.grid(axis='y', alpha=0.4)

# 显示图例
# plt.legend()
# 显示图例
plt.legend(
    loc='lower center',
    bbox_to_anchor=(0.5, 1.001),
    ncol=2,
    handlelength=0.5,
    handletextpad=0.5,
    frameon=False,
    fontsize=12
)

# 自动调整布局
plt.tight_layout()
# plt.savefig('Number of OpenAI API related question posts overtime.png', dpi=600)

# 保存为高清PDF文件
plt.savefig('Number_of_OpenAI_API_Related_Question_Posts.pdf', format='pdf', dpi=600, bbox_inches='tight')
plt.show()
```
